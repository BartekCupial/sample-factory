```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
discount = 0.99
```

```python
for discount in [0.9999, 0.99999, 1.0]:
    plt.plot([np.prod(np.ones(i) * discount) for i in range(128)], label=discount)
# plt.ylim(0, 1)
plt.legend()
```

```python
for gae_lambda in [0.95, 0.973, 0.985, 0.992, 0.996, 0.9985, 1.0]:
    plt.plot([np.prod(np.ones(i) * gae_lambda) for i in range(256)], label=gae_lambda)
plt.ylim(0, 1)
plt.legend()
```
